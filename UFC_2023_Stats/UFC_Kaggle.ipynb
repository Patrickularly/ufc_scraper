```python
weight_class_dict = {
                    'Strawweight':52.2,
                    'Flyweight': 56.7,
                    'Bantamweight': 61.2,
                    'Featherweight': 65.8,
                    'Lightweight': 70.3,
                    'Super lightweight': 74.8,
                    'Welterweight': 77.1,
                    'Super welterweight': 79.4,
                    'Middleweight': 83.9,
                    'Super middleweight': 88.5,
                    'Light heavyweight': 93.0,
                    'Cruiserweight': 102.1,
                    'Heavyweight': 120.2,
                    'Super heavyweight': 120.3
}

value_to_weight_class = {value: key for key, value in weight_class_dict.items()} 
value_to_weight_class
```

```python
def weight_class_mapper(weight):
    for key in value_to_weight_class.keys():
        if weight <= key:
            return value_to_weight_class[key]
    return 'Super heavyweight'
```

```python
df['weight_class'] = df['weight_in_kg'].apply(weight_class_mapper)
df
```

```python
df.date_of_birth = pd.to_datetime(df.date_of_birth)
```
